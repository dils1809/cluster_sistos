### Resultados de Exploración

Con la exploración realizada, se observó que el dataset contiene información sobre tráfico web, como URLs, hosts y códigos de respuesta. Con el cual se puede analizar patrones de tráfico web, identificar recursos más utilizados y evaluar la calidad del servicio mediante códigos de respuesta.

#### Tamaño del dataset: total de registros = 2,965,561

Es un dataset grande (casi 3 millones de registros), justifica el uso distribuido con Spark

#### Estructura de los datos 

Columnas:
- _c0 - Columna de índice
- host - cliente (IP o dominio)
- time - timestamp
- method - GET/POST
- url - recurso solicitado
- response - código HTTP
- bytes - tamaño de respuesta

El dataset presenta un problema estructural, ya que contiene una columna vacía adicional, lo que genera una columna _c0 innecesaria.

#### Estadísticas

- mean(response): ~209
- min(response): 200
- max(response): 501

Con lo anterior se puede interpretar que la mayoría de respuestas son exitosas (200) y que existen errores (400, 500)

Algunas columnas requieren conversión de tipos y manejo de valores especiales como "-" en bytes
